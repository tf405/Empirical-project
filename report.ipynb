```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

def scrape_leeds_players():
    # Create a new Safari session
    driver = webdriver.Safari()

    # Define the URL to scrape
    url = "https://www.transfermarkt.co.uk/leeds-united/kader/verein/399/saison_id/2017"
    
    # Open the URL
    driver.get(url)

    try:
        # Find all player rows
        player_rows = driver.find_elements(By.CSS_SELECTOR, '#yw1 > table > tbody > tr')
        
        # Initialize empty lists to store player names and values
        player_names = []
        player_values = []
        
        # Iterate over each player row
        for row in player_rows:
            try:
                # Find the player name element within the current row
                player_name_elem = row.find_element(By.CSS_SELECTOR, 'td:nth-child(2) a')
                
                # Get the player name text
                player_name = player_name_elem.text.strip()
                
                # Find the player value element within the current row
                player_value_elem = row.find_element(By.CSS_SELECTOR, 'td.rechts.hauptlink')
                
                # Get the player value text
                player_value = player_value_elem.text.strip()
                
                # Append player name and value to the respective lists
                player_names.append(player_name)
                player_values.append(player_value)
            except Exception as e:
                print("Error while extracting data from player row:", e)
        
        # Create a DataFrame from the lists
        df = pd.DataFrame({'Player Name': player_names, 'Player Value': player_values})
        
        return df
        
    except Exception as e:
        print("Error while scraping:", e)
    finally:
        # Close the browser session
        driver.quit()

# Call the function to scrape Leeds United players and store data in a DataFrame
leeds_players_df17 = scrape_leeds_players()
print(leeds_players_df17)

```

```python
import matplotlib.pyplot as plt

# Convert string values to numerical format
df_income.iloc[:, 1:] = df_income.iloc[:, 1:].replace('[\$,]', '', regex=True).astype(float)

# Exclude 'Total' category
df_income = df_income.drop(columns=['Total'])

# Data
years = df_income['Year']
categories = df_income.columns[1:]  # Exclude 'Year' column

# Plotting for years 2017 and 2018
plt.figure(figsize=(10, 6))
df_income[df_income['Year'].isin([2017, 2018])].plot(x='Year', kind='bar', stacked=True, width=0.8, ax=plt.gca())  # Use the current axis
plt.xlabel('Year')
plt.ylabel('Income (£10 million)')
plt.title('Income Breakdown by Year and Category (2017-2018)')
plt.legend(loc='upper left')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))  # Use scientific notation for y-axis
plt.tight_layout()
plt.show()

# Plotting for years 2021 and 2022
plt.figure(figsize=(10, 6))
df_income[df_income['Year'].isin([2021, 2022])].plot(x='Year', kind='bar', stacked=True, width=0.8, ax=plt.gca())  # Use the current axis
plt.xlabel('Year')
plt.ylabel('Income (£100 million)')
plt.title('Income Breakdown by Year and Category (2021-2022)')
plt.legend(loc='upper left')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))  # Use scientific notation for y-axis
plt.tight_layout()
plt.show()
```
